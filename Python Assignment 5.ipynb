#### Write a Python program to print the first 8 terms of an arithmetic progression starting
with 3 and having a common difference of 4.
The program should output the following sequence:
3 7 11 15 19 23 27 31 

```python
a = 3      # 1st term
d = 4      # common difference
n = 8      # number of terms

print("The first 8 terms of the arithmetic progression are:")

for i in range(n):
    term = a + i * d
    print(term, end=" ")
```

####  Write a Python program to print the first 6 terms of a geometric sequence starting
with 2 and having a common ratio of 3.
The program should output the following sequence:
2 6 18 54 162 486 

```python
a = 2      
r = 3      # common ratio
n = 6      

print("The first 6 terms of the geometric sequence are:")

for i in range(n):
    term = a * (r ** i)
    print(term, end=" ")
```

#### Write a program that asks the user for a positive integer value. The program should
calculate the sum of all the integers from 1 up to the number entered. For example, if
the user enters 20, the loop will find the sum of 1, 2, 3, 4, ... 20. 

```python
n = int(input("Enter a positive integer: "))

sum_numbers = 0

for i in range(1, n + 1):
    sum_numbers += i

print(f"The sum of numbers from 1 to {n} is: {sum_numbers}")
```

#### . write a program that takes a positive integer N as input and calculates the sum ofthe reciprocals of all numbers from 1 up to N. The program should display the final sum.Output of the program should be like: Enter a positive integer: 5 The sum of reciprocals from 1 to 5 is: 2.28 

```python
N = int(input("Enter a positive integer: "))

sum_reciprocal = 0

for i in range(1, N + 1):
    sum_reciprocal += 1 / i

print(f"The sum of reciprocals from 1 to {N} is: {sum_reciprocal:.2f}")
```

#### Write a program that prompts the user to enter a positive integer and calculates its
factorial. The factorial of a positive integer 'n' is denoted as 'n!' and is calculated by
multiplying all the integers from 1 to 'n' together. For example, the factorial of 5
(denoted as 5!) is calculated as 1 x 2 x 3 x 4 x 5.
The program should display the factorial value if the input is a positive number, or
display a message stating that the factorial does not exist for negative numbers.
Additionally, for an input of zero, the program should output that the factorial of 0 is 1. 

```python
n = int(input("Enter a number: "))

if n < 0:
    print("Sorry, factorial does not exist for negative numbers.")
elif n == 0:
    print("The factorial of 0 is 1.")
else:
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i  
    print(f"The factorial of {n} is {factorial}.")
```

#### Write a Python program that prompts the user to enter a base number and an
exponent, and then calculates the power of the base to the exponent. The program
should not use the exponentiation operator (**) or the math.pow() function. The
program should handle both positive and negative exponents. 

```python

base = float(input("Enter the base number: "))
exp = int(input("Enter the exponent: "))

result = 1

if exp > 0:
    for i in range(exp):
        result *= base
elif exp < 0:
    for i in range(-exp):
        result *= base
    result = 1 / result
else:
    result = 1  

print(f"{base} raised to the power {exp} is: {result}")
```
