#### Write a program that accepts a list from user and print the alternate element of list.

#### Write a program that accepts a list from user. Your program should reverse the content of list and display it. Do not use reverse() method.

```python
lst = input("Enter elements of list separated by space: ").split()
reversed_list = []
for i in range(len(lst)-1, -1, -1):
    reversed_list.append(lst[i])
print("Reversed list:", reversed_list)
```

#### Find and display the largest number of a list without using built-in function max(). Your program should ask the user to input values in list from keyboard.

```python
lst = list(map(int, input("Enter numbers separated by space: ").split()))
largest = lst[0]
for num in lst:
    if num > largest:
        largest = num
print("Largest number is:", largest)
```

#### Write a program that rotates the element of a list so that the element at the first index moves to the second index, the element in the second index moves to the third index, etc., and the element in the last index moves to the first index.

```python
lst = input("Enter elements of list separated by space: ").split()
rotated = lst[-1:] + lst[:-1]   
print("Rotated list:", rotated)
```

#### Write a program that reads a string from the user containing a date in the form mm/dd/yyyy. It should print the date in the form March 12, 2021.

```python
months = ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December"]

date = input("Enter date (mm/dd/yyyy): ")
mm, dd, yyyy = date.split("/")
mm = int(mm)
print(f"{months[mm-1]} {int(dd)}, {yyyy}")
```

#### Write a program with a function that accepts a string from keyboard and create a new string after converting character of each word capitalized. For instance, if the sentence is "stop and smell the roses." the output should be "Stop And Smell The Roses"Write a program with a function that accepts a string from keyboard and create a new string after converting character of each word capitalized. For instance, if the sentence is "stop and smell the roses." the output should be "Stop And Smell The Roses"

```python
def capitalize_words(sentence):
    words = sentence.split()
    new_words = []
    for w in words:
        new_words.append(w.capitalize())
    return " ".join(new_words)

s = input("Enter a sentence: ")
print(capitalize_words(s))
```

#### Write a program to add two matrices of size n x m.

```python
n = int(input("Enter number of rows: "))
m = int(input("Enter number of columns: "))

print("Enter first matrix:")
A = [[int(x) for x in input().split()] for _ in range(n)]

print("Enter second matrix:")
B = [[int(x) for x in input().split()] for _ in range(n)]

C = [[A[i][j] + B[i][j] for j in range(m)] for i in range(n)]

print("Resultant Matrix (A+B):")
for row in C:
    print(row)
```

####  Write a program to multiply two matrices

```python
n = int(input("Enter rows for first matrix: "))
m = int(input("Enter columns for first matrix (and rows for second): "))
p = int(input("Enter columns for second matrix: "))

print("Enter first matrix:")
A = [[int(x) for x in input().split()] for _ in range(n)]

print("Enter second matrix:")
B = [[int(x) for x in input().split()] for _ in range(m)]

# Multiplication result matrix
C = [[0 for _ in range(p)] for _ in range(n)]

for i in range(n):
    for j in range(p):
        for k in range(m):
            C[i][j] += A[i][k] * B[k][j]

print("Resultant Matrix (A*B):")
for row in C:
    print(row)
```
